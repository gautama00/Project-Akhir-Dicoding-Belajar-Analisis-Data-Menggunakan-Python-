### Analisis Lanjutan : RFM Analysis

RFM Analisis dilakukan untuk meninjau segmentasi customer pemakai platform ini. Segmentasi dilakukan berdasarkan waktu terakhir kali bertransaksi, frequensi transaksi, dan nominal transaksu yang dilakukan. Analisis ini memungkinkan bisnis untuk mengidentifikasi pelanggan bernilai tinggi, mempertahankan pelanggan loyal, mengoptimalkan strategi pemasaran yang spesifik, serta meningkatkan efisiensi alokasi sumber daya melalui wawasan berbasis data, sehingga mendukung pertumbuhan bisnis yang berkelanjutan dan peningkatan retensi pelanggan.

```python
rfm_df = alldata.groupby(by="customer_unique_id", as_index=False).agg({
    "order_approved_at": "max", #mengambil tanggal order terakhir
    "order_id": "nunique",
    "payment_value": "sum"
})
rfm_df.columns = ["customer_unique_id", "max_order_timestamp", "frequency", "monetary"]
rfm_df.head()
```

```python
#cek nilai kosong
print(rfm_df.max_order_timestamp.isna().sum(),
      rfm_df.frequency.isna().sum(),
      rfm_df.monetary.isna().sum())
```

```python
#hapus baris dengan nilai kosong
rfm_df = rfm_df.dropna(subset=["max_order_timestamp"])
print("data kosong max_order_timestamp = ",rfm_df.max_order_timestamp.isna().sum())
```

```python
rfm_df["max_order_timestamp"] =  pd.to_datetime(rfm_df["max_order_timestamp"])
```

```python
alldata.order_approved_at.isna().sum()
```

```python
alldata_rfm = alldata[alldata["order_approved_at"].isna()==False]
alldata_rfm.order_approved_at.isna().sum()
```

```python
print(rfm_df["max_order_timestamp"].dtype)
print(alldata_rfm["order_approved_at"].dtype)
```

```python
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = alldata_rfm["order_approved_at"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)
rfm_df.head()
```

```python
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(6, 15))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(x="recency", y="customer_unique_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)

sns.barplot(x="frequency", y="customer_unique_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

sns.barplot(x="monetary", y="customer_unique_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15, rotation = 25)

plt.suptitle("Best Customer Based on RFM Parameters (customer_unique_id)", fontsize=20)
plt.show()
```

**Mengurustkan peringkat customer berdasarkan frequency, recency, dan monetary score**

```python
rfm_df['r_rank'] = rfm_df['recency'].rank(ascending=False)
rfm_df['f_rank'] = rfm_df['frequency'].rank(ascending=True)
rfm_df['m_rank'] = rfm_df['monetary'].rank(ascending=True)

rfm_df.head()
     
```

```python
# normalizing the rank of the customers
rfm_df['r_rank_norm'] = (rfm_df['r_rank']/rfm_df['r_rank'].max())*100
rfm_df['f_rank_norm'] = (rfm_df['f_rank']/rfm_df['f_rank'].max())*100
rfm_df['m_rank_norm'] = (rfm_df['m_rank']/rfm_df['m_rank'].max())*100
 
rfm_df.drop(columns=['r_rank', 'f_rank', 'm_rank'], inplace=True)
 
rfm_df.head()
```

```python
rfm_df['RFM_score'] = 0.15*rfm_df['r_rank_norm']+0.28 * \
    rfm_df['f_rank_norm']+0.57*rfm_df['m_rank_norm']
rfm_df['RFM_score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['customer_unique_id', 'RFM_score']].head(7)
```

**Segmentasi customer berdasarkan RFM score**

```python
rfm_df["customer_segment"] = np.where(
    rfm_df['RFM_score'] > 4.5, "Top customers", (np.where(
        rfm_df['RFM_score'] > 4, "High value customer",(np.where(
            rfm_df['RFM_score'] > 3, "Medium value customer", np.where(
                rfm_df['RFM_score'] > 1.6, 'Low value customers', 'lost customers'))))))

rfm_df[['customer_unique_id', 'RFM_score', 'customer_segment']].head(20)
```

```python
customer_segment_df = rfm_df.groupby(by="customer_segment", as_index=False).customer_unique_id.nunique()
customer_segment_df
```

```python
customer_segment_df['customer_segment'] = pd.Categorical(customer_segment_df['customer_segment'], [
    "lost customers", "Low value customers", "Medium value customer",
    "High value customer", "Top customers"
])
```

```python
customer_segment_df.sort_values(by="customer_segment", ascending=True)
```
